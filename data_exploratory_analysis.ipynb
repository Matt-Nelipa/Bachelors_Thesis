импортируем необходимые библиотеки

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

импортируем все доступные нам данные

```python
events_df = pd.read_csv('data/Retailrocket_Dataset/events.csv')
ip_part_1_df = pd.read_csv('data/Retailrocket_Dataset/item_properties_part1.csv')
ip_part_2_df = pd.read_csv('data/Retailrocket_Dataset/item_properties_part2.csv')
ct_df = pd.read_csv('data/Retailrocket_Dataset/category_tree.csv')
```

```python
ip_df = pd.concat([ip_part_1_df, ip_part_2_df], ignore_index=True)
```
